```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Guna Numpy
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots() #Buat kanvas
ax.plot(x, y1, color='blue', label='Sin')
ax.plot(x, y2, color='red', label='Cos')

ax.set_title("Perbandingan Sin & Cos")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

nama_buah = ["Epal", "Oren", "Lemon"]
jualan = [10, 15, 5]

warna = ["#ff6b6b", "#ffa94d", "#ffd43b"]

plt.figure(figsize=(8,5))
bar = plt.bar(nama_buah, jualan, color=warna)

plt.title("Graf Jualan Buah", fontsize=16, fontweight="bold")
plt.xlabel("Nama Buah")
plt.ylabel("Bilangan Jualan")

for i in bar:
    plt.text(
        i.get_x() + i.get_width()/2,
        i.get_height() + 0.2,
        str(int(i.get_height())),
        ha='center',
        fontsize=11,
        fontweight='bold'
    )

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# ==============================================================================
# 📊 SISTEM PENGURUSAN DAN KIRA MARKAH PELAJAR
# Kursus: Asas Sains Komputer / Pengaturcaraan Python
# Disediakan oleh: [Nama Anda]
# No. Matrik:     [No. Matrik Anda]
# ==============================================================================

def sistem_kira_markah():
    # Banner hiasan untuk skrin input
    print("=" * 50)
    print("         SISTEM KIRA MARKAH & GRED AKADEMIK       ")
    print("=" * 50)

    # 1. Mengambil Input Data Pelajar
    nama_pelajar = input("▶️ Masukkan Nama Pelajar : ")

    try:
        # 2. Mengambil Input Markah (Guna float untuk perpuluhan)
        math = float(input("▶️ Markah Matematik (0-100) : "))
        sains = float(input("▶️ Markah Sains (0-100)     : "))
        english = float(input("▶️ Markah English (0-100)   : "))

        # Validasi ralat sekiranya markah luar dari julat 0-100
        if not (0 <= math <= 100 and 0 <= sains <= 100 and 0 <= english <= 100):
            print("\n❌ Ralat: Markah mestilah di antara julat 0 hingga 100 sahaja.")
            return

    except ValueError:
        # Menghalang program daripada 'crash' jika pensyarah tersalah taip huruf
        print("\n❌ Ralat: Sila masukkan nilai nombor yang sah untuk markah.")
        return

    # 3. Pengiraan Jumlah dan Purata Markah
    jumlah_markah = math + sains + english
    purata = jumlah_markah / 3

    # 4. Penentuan Gred & Status Menggunakan Logik 'if-elif-else'
    if purata >= 80:
        gred = "A (Cemerlang)"
        status = "LULUS (Anugerah Dekan)"
    elif purata >= 60:
        gred = "B (Kepujian)"
        status = "LULUS"
    elif purata >= 50:
        gred = "C (Lulus)"
        status = "LULUS"
    else:
        gred = "F (Gagal)"
        status = "GAGAL (Perlu Perhatian)"

    # 5. Paparan Slip Keputusan Rasmi (Ini bahagian paling pensyarah suka)
    print("\n" + "=" * 50)
    print(f"📄 SLIP KEPUTUSAN RASMI: {nama_pelajar.upper()}")
    print("=" * 50)
    print(f"🔹 Total Markah     : {jumlah_markah:.1f} / 300.0")
    print(f"🔹 Purata Markah    : {purata:.2f}%")
    print(f"🔹 Gred Keseluruhan : {gred}")
    print("-" * 50)
    print(f"📢 STATUS AKHIR     : {status}")
    print("=" * 50)

# Menjalankan fungsi sistem secara automatik
sistem_kira_markah()
```

```python
# ==============================================================================
# 📊 SISTEM KIRA MARKAH INTERAKTIF (GUI INTERFACE)
# Kursus: Asas Sains Komputer / Pengaturcaraan Python
# Disediakan oleh: [Nama Anda]
# No. Matrik:     [No. Matrik Anda]
# ==============================================================================

import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Cipta komponen Interface (Widgets)
output_area = widgets.Output()

txt_nama = widgets.Text(description="Nama Pelajar:", placeholder="Masukkan nama...")
num_math = widgets.BoundedFloatText(description="Matematik:", min=0, max=100, value=0)
num_sains = widgets.BoundedFloatText(description="Sains:", min=0, max=100, value=0)
num_english = widgets.BoundedFloatText(description="English:", min=0, max=100, value=0)

btn_kira = widgets.Button(description="Kira & Jana Slip", button_style="success", icon="calculator")

# 2. Fungsi Logik & Pengiraan Gred
def on_button_clicked(b):
    with output_area:
        clear_output() # Padam keputusan lama sebelum keluar keputusan baru

        nama = txt_nama.value.strip()
        math = num_math.value
        sains = num_sains.value
        english = num_english.value

        if not nama:
            print("❌ Ralat: Sila masukkan Nama Pelajar terlebih dahulu!")
            return

        # Kira Jumlah & Purata
        jumlah = math + sains + english
        purata = jumlah / 3

        # Logik Gred & Status
        if purata >= 80:
            gred, status = "A (Cemerlang)", "LULUS (Anugerah Dekan)"
        elif purata >= 60:
            gred, status = "B (Kepujian)", "LULUS"
        elif purata >= 50:
            gred, status = "C (Lulus)", "LULUS"
        else:
            gred, status = "F (Gagal)", "GAGAL (Perlu Perhatian)"

        # Paparan Output Slip Keputusan yang Kemas
        print("=" * 50)
        print(f"📄 SLIP KEPUTUSAN RASMI: {nama.upper()}")
        print("=" * 50)
        print(f"🔹 Total Markah     : {jumlah:.1f} / 300.0")
        print(f"🔹 Purata Markah    : {purata:.2f}%")
        print(f"🔹 Gred Keseluruhan : {gred}")
        print("-" * 50)
        print(f"📢 STATUS AKHIR     : {status}")
        print("=" * 50)

# 3. Hubungkan butang dengan fungsi logik
btn_kira.on_click(on_button_clicked)

# 4. Susun & Paparkan Interface di Google Colab
print("" * 50)
print("💻 SILA MASUKKAN DATA PADA INTERFACE DI BAWAH:")
print("-" * 50)
display(txt_nama, num_math, num_sains, num_english, btn_kira, output_area)
```
