```python
# 1. Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 2. Sediakan kanvas lukisan (Figure) tanpa bingkai putih yang mengganggu
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis('off') # Sorok paksi X dan Y supaya nampak macam kartun sebenar

# 3. Lukis Latar Belakang (Langit, Rumput, Matahari, Awan)
ax.set_facecolor('#87CEEB') # Warna langit biru
fig.patch.set_facecolor('#87CEEB')

# Rumput hijau
ax.fill_between([0, 10], 0, 1.5, color='#4CAF50')

# Matahari kuning comel
matahari = plt.Circle((8.5, 4.5), 0.5, color='#FFD700')
ax.add_patch(matahari)

# Objek Awan yang akan bergerak
awan, = ax.plot([], [], color='white', marker='o', markersize=25, linestyle='None', alpha=0.9)
awan_x = [2, 3, 6, 7]
awan_y = [4.5, 4.6, 5.0, 5.1]

# 4. Lukis Bahagian Badan Dinosaur (Menggunakan koordinat titik)
# Badan T-Rex (Warna Hijau Comel)
badan = plt.Polygon([[4, 2], [5.2, 2.5], [5.5, 3.5], [5.0, 4.2], [4.2, 4.2], [3.8, 3.5], [3.2, 2.3]], color='#8BC34A')
ax.add_patch(badan)

# Mata dinosaur
mata = plt.Circle((4.4, 3.9), 0.08, color='black')
ax.add_patch(mata)

# Tangan kecil T-Rex
tangan, = ax.plot([4.8, 5.1], [3.2, 3.1], color='#7CB342', linewidth=5, solid_capstyle='round')

# Kaki dinosaur yang akan bergerak (Kaki Kiri & Kaki Kanan)
kaki_kiri, = ax.plot([], [], color='#689F38', linewidth=8, solid_capstyle='round')
kaki_kanan, = ax.plot([], [], color='#558B2F', linewidth=8, solid_capstyle='round')

# Ekor dinosaur
ekor = plt.Polygon([[3.3, 2.4], [2.2, 2.8], [3.5, 3.0]], color='#8BC34A')
ax.add_patch(ekor)

# 5. Fungsi Animasi (Dipanggil berulang kali untuk gerakkan watak)
def kemaskini_animasi(i):
    # Gerakkan awan ke kanan secara perlahan, kalau terkeluar skrin dia patah balik
    awan_semasa_x = [(x + i * 0.02) % 11 - 1 for x in awan_x]
    awan.set_data(awan_semasa_x, awan_y)

    # Gerakan kaki menggunakan formula Matematik (Sinus & Kosinus) untuk efek berjalan
    ayunan = np.sin(i * 0.4) * 0.3

    # Kaki Kiri melangkah
    kaki_kiri.set_data([4.1, 4.1 + ayunan], [2.0, 1.2])

    # Kaki Kanan melangkah berlawanan arah
    kaki_kanan.set_data([4.7, 4.7 - ayunan], [2.0, 1.2])

    # Sedikit kesan badan terhenjut-henjut comel semasa berjalan
    henjut = np.abs(np.sin(i * 0.4)) * 0.05
    badan.set_xy([[4, 2+henjut], [5.2, 2.5+henjut], [5.5, 3.5+henjut], [5.0, 4.2+henjut], [4.2, 4.2+henjut], [3.8, 3.5+henjut], [3.2, 2.3+henjut]])
    mata.set_center((4.4, 3.9 + henjut))
    tangan.set_data([4.8, 5.1], [3.2 + henjut, 3.1 + henjut])
    ekor.set_xy([[3.3, 2.4+henjut], [2.2, 2.8+henjut], [3.5, 3.0+henjut]])

    return awan, kaki_kiri, kaki_kanan, badan, mata, tangan, ekor

# 6. Jana Animasi (60 frame untuk pergerakan yang lancar)
animasi = FuncAnimation(fig, kemaskini_animasi, frames=60, interval=50, blit=True)

# Padamkan plot statik asal supaya tidak bertindan
plt.close()

# 7. Paparkan video animasi kartun di Google Colab!
HTML(animasi.to_html5_video())
```

```python
from IPython.display import HTML, display

# Kod gabungan Python + HTML5 Canvas + JavaScript untuk animasi perang & zombie pelbagai variasi
game_war_kod = """
<div id="game-container" style="text-align: center; font-family: 'Courier New', Courier, monospace; background-color: #1a1a1a; padding: 10px; color: white; width: 820px; margin: 0 auto; border: 4px solid #555; box-shadow: 0 0 20px rgba(255, 0, 0, 0.3);">
    <h2 style="color: #ff3333; margin: 5px;">🔥 ZOMBIE WARZONE: FRONTPLAY COLAB 🔥</h2>
    <p style="margin: 2px;">Medan Perang: Klik Zombie Untuk Tembak | Tekan <b>'R'</b> Untuk Reload</p>

    <canvas id="gameCanvas" width="800" height="400" style="background: #111; cursor: crosshair; border: 2px solid #ff3333;"></canvas>
</div>

<script>
    const canvas = document.getElementById("gameCanvas");
    const ctx = canvas.getContext("2d");

    // Suara sintetik perang
    function playSound(type) {
        try {
            let context = new (window.AudioContext || window.webkitAudioContext)();
            let osc = context.createOscillator();
            let gain = context.createGain();
            osc.connect(gain);
            gain.connect(context.destination);

            if (type === 'shoot') {
                osc.type = 'sawtooth';
                osc.frequency.setValueAtTime(400, context.currentTime);
                osc.frequency.exponentialRampToValueAtTime(20, context.currentTime + 0.15);
                gain.gain.setValueAtTime(0.4, context.currentTime);
                gain.gain.linearRampToValueAtTime(0.01, context.currentTime + 0.15);
                osc.start(); osc.stop(context.currentTime + 0.15);
            } else if (type === 'reload') {
                osc.type = 'triangle';
                osc.frequency.setValueAtTime(200, context.currentTime);
                osc.frequency.setValueAtTime(500, context.currentTime + 0.1);
                gain.gain.setValueAtTime(0.3, context.currentTime);
                gain.gain.linearRampToValueAtTime(0.01, context.currentTime + 0.25);
                osc.start(); osc.stop(context.currentTime + 0.25);
            } else if (type === 'hit') {
                osc.type = 'square';
                osc.frequency.setValueAtTime(60, context.currentTime);
                gain.gain.setValueAtTime(0.5, context.currentTime);
                gain.gain.linearRampToValueAtTime(0.01, context.currentTime + 0.2);
                osc.start(); osc.stop(context.currentTime + 0.2);
            }
        } catch(e) {}
    }

    // Gambar Zombie Sebenar & Gambar Latar Belakang Perang
    const zombieImg = new Image();
    zombieImg.src = "https://images.unsplash.com/photo-1509248961158-e54f6934749c?w=120&auto=format&fit=crop&q=60";

    // Variabel Game
    let playerHp = 100;
    let score = 0;
    let wave = 1;
    let ammo = 10; // Extra peluru sebab perang!
    let maxAmmo = 10;
    let isReloading = false;
    let zombies = [];
    let particles = [];
    let environmentParticles = [];
    let gameOver = false;
    let lastSpawnTime = 0;

    // Cipta efek debu, asap, dan abu perang (Ambience)
    for (let i = 0; i < 30; i++) {
        environmentParticles.push({
            x: Math.random() * canvas.width,
            y: Math.random() * canvas.height,
            size: Math.random() * 3 + 1,
            speedX: -(Math.random() * 0.8 + 0.2), // Tiup ke kiri
            speedY: (Math.random() - 0.5) * 0.5,
            alpha: Math.random() * 0.5 + 0.2
        });
    }

    // Kelas Zombie dengan Gaya Berjalan "War Situation"
    class WarZombie {
        constructor() {
            this.reset();
            this.x = canvas.width + Math.random() * 150; // Mula dari luar skrin
        }

        reset() {
            this.x = canvas.width + Math.random() * 200;

            // Jenis Pergerakan Semasa Perang (Dipilih secara rawak)
            // 1. RUNNER (Laju, kurus/kecil)
            // 2. LIMPER (Terhincut-hincut, saiz normal, jalan guna fungsi Sinus)
            // 3. CRAWLER (Merangkak di bawah, perlahan tapi liat)
            const types = ["runner", "limper", "crawler"];
            this.type = types[Math.floor(Math.random() * types.length)];

            if (this.type === "runner") {
                this.y = Math.random() * (canvas.height - 150) + 50;
                this.baseSpeed = 2.2 * (1 + (wave * 0.1));
                this.size = 45;
                this.hp = Math.ceil(wave / 2);
            } else if (this.type === "limper") {
                this.y = Math.random() * (canvas.height - 150) + 50;
                this.baseSpeed = 1.0 * (1 + (wave * 0.1));
                this.size = 60;
                this.hp = Math.ceil(wave * 1);
                this.limpTimer = 0;
            } else { // Crawler
                this.y = canvas.height - 75; // Sentiasa merangkak di tanah
                this.baseSpeed = 0.6 * (1 + (wave * 0.1));
                this.size = 55;
                this.hp = Math.ceil(wave * 1.5); // Lebih liat
            }

            this.initialY = this.y;
        }

        update() {
            // LOGIK BERJALAN SITUASI WAR
            if (this.type === "runner") {
                // Berlari laju meluru ke hadapan sambil bergegar ketakutan/marah
                this.x -= this.baseSpeed;
                this.y = this.initialY + (Math.random() - 0.5) * 4;
            }
            else if (this.type === "limper") {
                // Jalan terhincut-hincut (Laju sekejap, lambat sekejap mengikut hayunan kaki cedera)
                this.limpTimer += 0.15;
                let speedModifier = Math.sin(this.limpTimer) > 0 ? 2.0 : 0.2;
                this.x -= this.baseSpeed * speedModifier;

                // Badan terjungkit-jungkit ke atas dan bawah semasa menghincut
                this.y = this.initialY + (Math.sin(this.limpTimer) * 8);
            }
            else if (this.type === "crawler") {
                // Merangkak perlahan di atas tanah dengan gerakan heretan kiri ke kanan
                this.x -= this.baseSpeed;
                this.y = (canvas.height - 75) + (Math.sin(this.x * 0.05) * 5);
            }

            // Jika sampai ke kubu pertahanan pemain (Hujung kiri)
            if (this.x < -this.size && !gameOver) {
                playerHp -= this.type === "runner" ? 10 : 20; // Crawler/Limper gigit lagi sakit
                playSound('hit');
                this.reset();
            }
        }

        draw() {
            ctx.save();

            // Beri warna/aura berbeza mengikut jenis ancaman perang
            if (this.type === "runner") {
                ctx.shadowColor = "#ff6600"; // Aura oren laju
            } else if (this.type === "limper") {
                ctx.shadowColor = "#00ff00"; // Aura hijau zombi toksik
            } else {
                ctx.shadowColor = "#ff0000"; // Aura merah crawler merbahaya
            }
            ctx.shadowBlur = 12;

            // Kesan khas untuk crawler (kelihatan leper/merangkak)
            if (this.type === "crawler") {
                if (zombieImg.complete) {
                    ctx.drawImage(zombieImg, this.x, this.y + 20, this.size + 15, this.size - 20);
                } else {
                    ctx.fillStyle = "#3a4d3f";
                    ctx.fillRect(this.x, this.y + 20, this.size + 15, this.size - 20);
                }
            } else {
                if (zombieImg.complete) {
                    ctx.drawImage(zombieImg, this.x, this.y, this.size, this.size);
                } else {
                    ctx.fillStyle = "#4A5D4E";
                    ctx.fillRect(this.x, this.y, this.size, this.size);
                }
            }
            ctx.restore();

            // Label jenis zombie di atas kepala untuk strategi perang
            ctx.fillStyle = "white";
            ctx.font = "9px Arial";
            ctx.fillText(this.type.toUpperCase(), this.x, this.y - 15);

            // Nyawa Zombie
            ctx.fillStyle = "#444";
            ctx.fillRect(this.x, this.y - 10, this.size, 4);
            ctx.fillStyle = "red";
            ctx.fillRect(this.x, this.y - 10, this.size * (this.hp / (this.type === "runner" ? Math.ceil(wave/2) : (this.type === "limper" ? wave : Math.ceil(wave*1.5)))), 4);
        }
    }

    // Kelas Percikan Darah/Kesan Letupan Bom
    class WarParticle {
        constructor(x, y, color) {
            this.x = x;
            this.y = y;
            this.vx = (Math.random() - 0.5) * 8;
            this.vy = (Math.random() - 0.5) * 8;
            this.alpha = 1;
            this.color = color || "red";
        }
        update() {
            this.x += this.vx;
            this.y += this.vy;
            this.alpha -= 0.04;
        }
        draw() {
            ctx.save();
            ctx.globalAlpha = this.alpha;
            ctx.fillStyle = this.color;
            ctx.beginPath();
            ctx.arc(this.x, this.y, 3, 0, Math.PI * 2);
            ctx.fill();
            ctx.restore();
        }
    }

    function spawnZombie(timestamp) {
        let spawnRate = 1600 - (wave * 120);
        if (spawnRate < 500) spawnRate = 500;

        if (timestamp - lastSpawnTime > spawnRate) {
            if (zombies.length < 4 + wave) {
                zombies.push(new WarZombie());
            }
            lastSpawnTime = timestamp;
        }
    }

    // Tembakan Mouse Click
    canvas.addEventListener("mousedown", function(event) {
        if (gameOver || isReloading || ammo <= 0) {
            if (ammo <= 0 && !isReloading) playSound('hit');
            return;
        }

        const rect = canvas.getBoundingClientRect();
        const mouseX = event.clientX - rect.left;
        const mouseY = event.clientY - rect.top;

        ammo--;
        playSound('shoot');

        // Cipta kilatan cahaya tembakan (Muzzle Flash) di skrin
        ctx.fillStyle = "rgba(255, 255, 200, 0.4)";
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        for (let i = zombies.length - 1; i >= 0; i--) {
            let z = zombies[i];
            let actualY = z.type === "crawler" ? z.y + 20 : z.y;
            let actualHeight = z.type === "crawler" ? z.size - 20 : z.size;
            let actualWidth = z.type === "crawler" ? z.size + 15 : z.size;

            if (mouseX >= z.x && mouseX <= z.x + actualWidth &&
                mouseY >= actualY && mouseY <= actualY + actualHeight) {

                z.hp--;

                // Letupan darah percikan api medan perang
                for(let p=0; p<12; p++) particles.push(new WarParticle(mouseX, mouseY, "red"));
                for(let p=0; p<5; p++) particles.push(new WarParticle(mouseX, mouseY, "#ffcc00")); // Spark api

                if (z.hp <= 0) {
                    score += z.type === "runner" ? 15 : 10; // Runner bagi skor tinggi sebab laju
                    z.reset();
                    if (score % 60 === 0) wave++;
                }
                break;
            }
        }
    });

    // Butang Reload 'R'
    window.addEventListener("keydown", function(event) {
        if (event.key.toLowerCase() === 'r' && ammo < maxAmmo && !isReloading) {
            isReloading = true;
            playSound('reload');
            setTimeout(() => {
                ammo = maxAmmo;
                isReloading = false;
            }, 1000);
        }
    });

    // ENGINE RENDER UTAMA
    function drawWarzone(timestamp) {
        // Latar Belakang Perang (Langit merah gelap + bayang bangunan rosak)
        ctx.fillStyle = "#1c1313";
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        // Lukis bayangan siluet bandar perang/rosak di belakang
        ctx.fillStyle = "#0f0a0a";
        ctx.fillRect(50, 150, 80, 250);
        ctx.fillRect(200, 100, 120, 300);
        ctx.fillRect(450, 180, 90, 220);
        ctx.fillRect(600, 120, 100, 280);

        if (playerHp <= 0) gameOver = true;

        if (!gameOver) {
            spawnZombie(timestamp);

            // Gerakkan habuk & asap perang di udara
            environmentParticles.forEach(p => {
                p.x += p.speedX;
                p.y += p.speedY;
                if (p.x < 0) p.x = canvas.width;
                if (p.y < 0 || p.y > canvas.height) p.y = Math.random() * canvas.height;

                ctx.fillStyle = `rgba(150, 120, 110, ${p.alpha})`;
                ctx.beginPath();
                ctx.arc(p.x, p.y, p.size, 0, Math.PI * 2);
                ctx.fill();
            });

            // Update & Draw Zombis
            zombies.forEach(z => {
                z.update();
                z.draw();
            });

            // Update & Draw Efek Tembakan
            particles = particles.filter(p => p.alpha > 0);
            particles.forEach(p => {
                p.update();
                p.draw();
            });
        }

        // --- INTERFACE HUD WAR ---
        // Garisan Sempadan Pertahanan (Kubu Anda)
        ctx.strokeStyle = "rgba(255, 0, 0, 0.4)";
        ctx.lineWidth = 4;
        ctx.beginPath(); ctx.moveTo(5, 0); ctx.lineTo(5, canvas.height); ctx.stroke();

        // 1. Bar Nyawa Pangkalan
        ctx.fillStyle = "#222";
        ctx.fillRect(20, 20, 200, 22);
        ctx.fillStyle = playerHp > 40 ? "#ff3333" : "#880000"; // Warna merah darah untuk tema perang
        ctx.fillRect(20, 20, playerHp * 2, 22);
        ctx.strokeStyle = "#fff";
        ctx.strokeRect(20, 20, 200, 22);
        ctx.fillStyle = "#fff";
        ctx.font = "bold 11px 'Courier New'";
        ctx.fillText(`KUBU PERTAHANAN: ${playerHp}%`, 30, 35);

        // 2. Ammo Indicator
        ctx.font = "bold 15px 'Courier New'";
        ctx.fillStyle = isReloading ? "#ff9900" : (ammo === 0 ? "#ff0000" : "#fff");
        let ammoText = isReloading ? "🔋 RELOADING WEAPON..." : `MAG: ${"⚡".repeat(ammo)} (${ammo}/${maxAmmo})`;
        ctx.fillText(ammoText, 20, 68);

        // 3. Stats
        ctx.fillStyle = "#ffcc00";
        ctx.fillText(`💀 KILLS: ${score/10}`, canvas.width - 160, 35);
        ctx.fillStyle = "#ff3333";
        ctx.fillText(`🚨 SEKTOR: ${wave}`, canvas.width - 160, 60);

        if (gameOver) {
            ctx.fillStyle = "rgba(0, 0, 0, 0.9)";
            ctx.fillRect(0, 0, canvas.width, canvas.height);
            ctx.fillStyle = "red";
            ctx.font = "bold 45px 'Courier New'";
            ctx.fillText("💥 KUBU TELAH BOLOS 💥", canvas.width/2 - 240, canvas.height/2 - 20);
            ctx.fillStyle = "white";
            ctx.font = "20px 'Courier New'";
            ctx.fillText(`Mata Perang Anda: ${score}`, canvas.width/2 - 110, canvas.height/2 + 25);
            ctx.fillText("Klik ikon 'Play' di Colab untuk mulakan semula operasi!", canvas.width/2 - 280, canvas.height/2 + 65);
            return;
        }

        requestAnimationFrame(drawWarzone);
    }

    requestAnimationFrame(drawWarzone);
</script>
"""

# Run game di skrin Colab
display(HTML(game_war_kod))
```
